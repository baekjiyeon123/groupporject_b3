```python
# YOLOv5 모델 불러오기 (로컬에서 학습된 모델 경로 지정)
model = torch.hub.load('./', 'custom', path='runs/train/exp9/weights/best.pt', source='local')
from deep_sort_realtime.deepsort_tracker import DeepSort
# YOLOv5 모델 불러오기 (학습된 모델 경로 지정)
model = torch.hub.load('./', 'custom', path='runs/train/exp9/weights/best.pt', source='local')
# DeepSORT 트래커 초기화
deepsort = DeepSort(max_age=30, n_init=3, nms_max_overlap=1.0, max_cosine_distance=0.4)
# 비디오 파일 열기
cap = cv2.VideoCapture("/home/piai/jk1.mov")
# 비디오 저장 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/home/piai/tracked_output.mp4', fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))
# 색상 정의
color_dict = {
    "Drowning": (0, 0, 255),  # 빨간색 (BGR 형식)
    "Swimming": (255, 0, 0)   # 파란색 (BGR 형식)
}
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # YOLOv5로 객체 감지 수행
    results = model(frame)
    # 감지된 객체가 있는지 확인 (results.pred 사용)
    if results.pred is not None and len(results.pred) > 0:  # 감지된 객체가 있는지 확인
        bboxes = []
        class_names = []
        class_colors = []
        for r in results.pred[0]:  # 첫 번째 결과만 사용
            # 텐서를 item()으로 변환하여 (x, y, w, h, conf) 추출
            x1, y1, x2, y2 = r[:4].tolist()  # 좌상단과 우하단 좌표를 리스트로 변환
            conf = r[4].item()  # 신뢰도도 item()으로 변환
            class_idx = int(r[5].item())  # 클래스 인덱스 (int로 변환)
            class_name = results.names[class_idx]  # 클래스 이름 가져오기
            width = x2 - x1
            height = y2 - y1
            bboxes.append([[x1, y1, width, height], conf])  # DeepSORT에 필요한 형식 [[x1, y1, w, h], conf]
            class_names.append(f"{class_name} {conf:.2f}")  # 클래스명과 신뢰도를 저장
            # 클래스에 맞는 색상 선택
            if class_name in color_dict:
                class_colors.append(color_dict[class_name])  # 정의된 색상
            else:
                class_colors.append((0, 255, 0))  # 기본값은 녹색
        # bboxes 값 확인
        print(f"bboxes: {bboxes}")
        # DeepSORT로 트래킹 수행
        tracks = deepsort.update_tracks(bboxes, frame=frame) if len(bboxes) > 0 else []
        # 트래킹된 객체 표시
        for i, track in enumerate(tracks):
            if track.is_confirmed() and track.time_since_update <= 1:
                track_id = track.track_id
                bbox = track.to_tlbr()
                # 트래킹된 객체와 YOLOv5 감지 결과를 비교
                if i < len(class_names):  # 인덱스 범위 확인
                    color = class_colors[i]  # 클래스에 맞는 색상 선택
                    # 객체 경계 상자 그리기
                    cv2.rectangle(frame, (int(bbox[0]), int(bbox[1])), (int(bbox[2]), int(bbox[3])), color, 2)
                    # 클래스 이름과 신뢰도를 표시
                    cv2.putText(frame, class_names[i], (int(bbox[0]), int(bbox[1])-10), cv2.FONT_HERSHEY_SIMPLEX, 0.75, color, 2)
    else:
        tracks = []  # 객체가 감지되지 않으면 빈 리스트로 초기화
    # 비디오 파일로 저장
    out.write(frame)
# 자원 해제
cap.release()
out.release()
```
