```python
# YOLOv5 ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸° (ë¡œì»¬ì—ì„œ í•™ìŠµëœ ëª¨ë¸ ê²½ë¡œ ì§€ì •)
model = torch.hub.load('./', 'custom', path='runs/train/exp9/weights/best.pt', source='local')
from deep_sort_realtime.deepsort_tracker import DeepSort
# YOLOv5 ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸° (í•™ìŠµëœ ëª¨ë¸ ê²½ë¡œ ì§€ì •)
model = torch.hub.load('./', 'custom', path='runs/train/exp9/weights/best.pt', source='local')
# DeepSORT íŠ¸ëž˜ì»¤ ì´ˆê¸°í™”
deepsort = DeepSort(max_age=30, n_init=3, nms_max_overlap=1.0, max_cosine_distance=0.4)
# ë¹„ë””ì˜¤ íŒŒì¼ ì—´ê¸°
cap = cv2.VideoCapture("/home/piai/jk1.mov")
# ë¹„ë””ì˜¤ ì €ìž¥ ì„¤ì •
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/home/piai/tracked_output.mp4', fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))
# ìƒ‰ìƒ ì •ì˜
color_dict = {
    "Drowning": (0, 0, 255),  # ë¹¨ê°„ìƒ‰ (BGR í˜•ì‹)
    "Swimming": (255, 0, 0)   # íŒŒëž€ìƒ‰ (BGR í˜•ì‹)
}
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # YOLOv5ë¡œ ê°ì²´ ê°ì§€ ìˆ˜í–‰
    results = model(frame)
    # ê°ì§€ëœ ê°ì²´ê°€ ìžˆëŠ”ì§€ í™•ì¸ (results.pred ì‚¬ìš©)
    if results.pred is not None and len(results.pred) > 0:  # ê°ì§€ëœ ê°ì²´ê°€ ìžˆëŠ”ì§€ í™•ì¸
        bboxes = []
        class_names = []
        class_colors = []
        for r in results.pred[0]:  # ì²« ë²ˆì§¸ ê²°ê³¼ë§Œ ì‚¬ìš©
            # í…ì„œë¥¼ item()ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ (x, y, w, h, conf) ì¶”ì¶œ
            x1, y1, x2, y2 = r[:4].tolist()  # ì¢Œìƒë‹¨ê³¼ ìš°í•˜ë‹¨ ì¢Œí‘œë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
            conf = r[4].item()  # ì‹ ë¢°ë„ë„ item()ìœ¼ë¡œ ë³€í™˜
            class_idx = int(r[5].item())  # í´ëž˜ìŠ¤ ì¸ë±ìŠ¤ (intë¡œ ë³€í™˜)
            class_name = results.names[class_idx]  # í´ëž˜ìŠ¤ ì´ë¦„ ê°€ì ¸ì˜¤ê¸°
            width = x2 - x1
            height = y2 - y1
            bboxes.append([[x1, y1, width, height], conf])  # DeepSORTì— í•„ìš”í•œ í˜•ì‹ [[x1, y1, w, h], conf]
            class_names.append(f"{class_name} {conf:.2f}")  # í´ëž˜ìŠ¤ëª…ê³¼ ì‹ ë¢°ë„ë¥¼ ì €ìž¥
            # í´ëž˜ìŠ¤ì— ë§žëŠ” ìƒ‰ìƒ ì„ íƒ
            if class_name in color_dict:
                class_colors.append(color_dict[class_name])  # ì •ì˜ëœ ìƒ‰ìƒ
            else:
                class_colors.append((0, 255, 0))  # ê¸°ë³¸ê°’ì€ ë…¹ìƒ‰
        # bboxes ê°’ í™•ì¸
        print(f"bboxes: {bboxes}")
        # DeepSORTë¡œ íŠ¸ëž˜í‚¹ ìˆ˜í–‰
        tracks = deepsort.update_tracks(bboxes, frame=frame) if len(bboxes) > 0 else []
        # íŠ¸ëž˜í‚¹ëœ ê°ì²´ í‘œì‹œ
        for i, track in enumerate(tracks):
            if track.is_confirmed() and track.time_since_update <= 1:
                track_id = track.track_id
                bbox = track.to_tlbr()
                # íŠ¸ëž˜í‚¹ëœ ê°ì²´ì™€ YOLOv5 ê°ì§€ ê²°ê³¼ë¥¼ ë¹„êµ
                if i < len(class_names):  # ì¸ë±ìŠ¤ ë²”ìœ„ í™•ì¸
                    color = class_colors[i]  # í´ëž˜ìŠ¤ì— ë§žëŠ” ìƒ‰ìƒ ì„ íƒ
                    # ê°ì²´ ê²½ê³„ ìƒìž ê·¸ë¦¬ê¸°
                    cv2.rectangle(frame, (int(bbox[0]), int(bbox[1])), (int(bbox[2]), int(bbox[3])), color, 2)
                    # í´ëž˜ìŠ¤ ì´ë¦„ê³¼ ì‹ ë¢°ë„ë¥¼ í‘œì‹œ
                    cv2.putText(frame, class_names[i], (int(bbox[0]), int(bbox[1])-10), cv2.FONT_HERSHEY_SIMPLEX, 0.75, color, 2)
    else:
        tracks = []  # ê°ì²´ê°€ ê°ì§€ë˜ì§€ ì•Šìœ¼ë©´ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¡œ ì´ˆê¸°í™”
    # ë¹„ë””ì˜¤ íŒŒì¼ë¡œ ì €ìž¥
    out.write(frame)
# ìžì› í•´ì œ
cap.release()
out.release()
```
